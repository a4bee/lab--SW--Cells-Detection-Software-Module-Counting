```python
from ultralytics import YOLO
import os
import cv2
import matplotlib.pyplot as plt
import shutil
```

```python
model = YOLO("algaex40-healthy-infected-vol14_100epoch.pt")
```

```python
dir_path = os.path.dirname(os.path.abspath("cells_count.ipynb"))
if(not os.path.exists(dir_path + "\cells_counted")): os.makedirs(dir_path + "\cells_counted")
if(os.path.exists(dir_path + "\\runs")): shutil.rmtree(dir_path + "\\runs", ignore_errors=True)
```

```python
file_name = "test_photo3_40.jpg"
results = model.predict(file_name, save=True)
result = results[0]
```

```python
count_algae = 0
count_infection = 0

for item in result.boxes:
    if(item.cls[0].item() == 0):
        count_algae +=1
    else: count_infection += 1
```

```python
volume = 0.1*0.1*0.1*3*4 # x, y, z, rows, columns (mm^3)
concentration_algae = int(count_algae / volume)
print(volume, concentration_algae)
```

```python
os.chdir(dir_path + "\\runs\detect\predict")
image = cv2.imread(file_name)

plt.imshow(image)
plt.show()
```

```python
# box = image[0:52, 0:275]
# box_parameters = box.shape
# for n in range(box_parameters[0]):
#     for y in range(box_parameters[1]):
#         box[n][y] = (0,0,0)

box1 = image[0:52, 0:280]
box1_parameters = box1.shape
for n in range(box1_parameters[0]):
    for y in range(box1_parameters[1]):
        box1[n][y] = (0, 0, 0)

box2 = image[52:90, 0:500] 
box2_parameters = box2.shape
for n in range(box2_parameters[0]):
    for y in range(box2_parameters[1]):
        box2[n][y] = (0, 0, 0)

image = cv2.putText(image, "Counted Infections:  " + str(count_infection), (5,45), cv2.FONT_HERSHEY_SIMPLEX , 0.75, (255, 255, 255), 2, cv2.LINE_AA)
if(count_algae < 10):
    image = cv2.putText(image, "Counted Algaes:     " + str(count_algae), (5,20), cv2.FONT_HERSHEY_SIMPLEX , 0.75, (255, 255, 255), 2, cv2.LINE_AA)
else: image = cv2.putText(image, "Counted Algaes:    " + str(count_algae), (5,20), cv2.FONT_HERSHEY_SIMPLEX , 0.75, (255, 255, 255), 2, cv2.LINE_AA)

image = cv2.putText(image, "Algae concentration: " + str(concentration_algae) + ( " per cubic mm"), (5,80), cv2.FONT_HERSHEY_SIMPLEX , 0.75, (255, 255, 255), 2, cv2.LINE_AA) 
```

```python
os.chdir(dir_path + "\cells_counted")
cv2.imwrite(file_name, image)
```
